```python
def define_name(input, directions):

    input=input.split(',')
    directions=directions.split(',')
    total = len(input)
    counter = 0
    index = 0
    while counter <=total:


        direction = directions[counter][0]
        space = directions[counter][1]

        if direction == 'R':
            index += int(space)
            if index >= total:
                index = total-1
        elif direction == 'L':
            index -= int(space)
            if index < 0:
                index = 0



        counter+=1
    print(input[index])



input = 'Ardenisis,Grimindor,Silirin,Quorthyn,Kalxeth,Xildar,Kharpyr,Hyracion,Thyrosdaros,Havoris'
directions = 'L9,R1,L5,R9,L1,R3,L2,R9,L9,R4,L6'
define_name(input, directions)
```
